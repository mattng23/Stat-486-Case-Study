```python
import pandas as pd

moon = pd.read_csv("full_moon.csv")

holidays = pd.read_csv("holidays.csv")

weather = pd.read_csv("weather.csv")

crimes = pd.read_csv("Crimes.csv")
 
```

```python
#Change Date to Date and Time for crimes

crimes[['Date', 'Time']] = crimes['Date'].str.split(' ', 1, expand=True)

date_index = crimes.columns.get_loc("Date")

# Insert the "Time" column right after the "Date" column
crimes.insert(date_index + 1, "Time", crimes.pop("Time"))
```

```python
# Define the format you want for the date
date_format = "%m/%d/%Y"

# Convert the "Date" column to datetime and then back to string with the desired format
moon['FullMoonDates'] = pd.to_datetime(moon['FullMoonDates']).dt.strftime(date_format)
holidays['Date'] = pd.to_datetime(holidays['Date']).dt.strftime(date_format)
weather['datetime'] = pd.to_datetime(weather['datetime']).dt.strftime(date_format)
crimes['Date'] = pd.to_datetime(crimes['Date']).dt.strftime(date_format)
```

```python
#Drop unnecessary column in crimes df

crimes.drop(columns=["Year"], inplace=True)
```

```python
#Drop unnecessary column in weather df

weather.drop(columns=["name"], inplace=True)
```

```python
#Add moon df

#Delete year in moon df

moon.drop("Year", axis=1, inplace=True)

moon['FullMoon'] = 1

moon.rename(columns={'FullMoonDates': 'Date'}, inplace=True)
```

```python
#Merge Moon and Crimes

master_df = pd.merge(crimes, moon, on="Date", how="left")

#Change values to 0 if no full moon

master_df['FullMoon'].fillna(0, inplace=True)

master_df['FullMoon'] = master_df['FullMoon'].astype(int)

```

```python
master_df[master_df['Date'] == '01/25/2005']
```

```python
#Add holidays df

master_df = pd.merge(master_df, holidays, on="Date", how="left")
```

```python
master_df.rename(columns={'Day of Week': 'Holiday Day of Week'}, inplace=True)
```

```python
#Add weather df

weather.rename(columns={'datetime': 'Date'}, inplace=True)


master_df = pd.merge(master_df, weather, on="Date", how="left")
```
